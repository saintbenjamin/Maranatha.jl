```julia
using Maranatha
```

```julia
ff(x) = exp(-x) * cos(6x)
bounds = (0.0, π)
ns = 8
rule = :gauss_p4
boundary = :LU_EXEX
result_string = "demo"
```

```julia
Maranatha.PlotTools.plot_quadrature_coverage_1d(
    ff, 
    bounds..., 
    ns;
    rule=rule,
    boundary=boundary,
    name=result_string,
    save_file=true
)
```
