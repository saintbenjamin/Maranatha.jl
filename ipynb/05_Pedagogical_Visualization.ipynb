# `Maranatha.jl` :: Pedagogical Visualization

This notebook visualizes how quadrature rules sample and
approximate an integral.

Unlike the previous notebooks, which focused on **numerical
analysis and extrapolation**, the goal here is purely
**pedagogical**: to illustrate how quadrature nodes and weights
interact with the integrand.

These visualizations help develop intuition about

- where quadrature nodes are placed
- how different rules approximate the integrand
- how the integral estimate is constructed from weighted samples

## Load the **`Maranatha.jl`** package

We first load the **`Maranatha.jl`** package that provides the
quadrature visualization tools.

```julia
using Maranatha
```

## Visualize quadrature coverage using a TOML configuration

The function `plot_quadrature_coverage_1d` can visualize how
a quadrature rule samples the integrand over the integration
domain.

In this first example we load the integration setup from
a TOML configuration file, similar to the Quick Start notebook.

The plot illustrates:

- the true integrand curve
- quadrature nodes used by the rule
- the weighted sampling used to approximate the integral

```julia
toml_path = "../samples/sample_1d.toml"

toml_dir = dirname(toml_path)

result_string = "lesson5_sample_1d"
save_path = joinpath(toml_dir, "jld2")

plot_quadrature_coverage_1d(
    toml_path;
    N=9,
    name=result_string,
    figs_dir=save_path,
    save_file=true
)
```

## Define a custom example directly in Julia

Instead of using a TOML configuration file, we can also
define the integrand and numerical parameters directly
in Julia.

Here we construct a more oscillatory integrand and choose
a B-spline interpolation rule for demonstration.

```julia
integrand(x) = exp(-x) * cos(6x)
bounds = (0.0, π)
ns = 24
rule = :bspline_interp_p4
boundary = :LU_ININ
result_string = result_string
save_file = true
```

## Visualize the quadrature rule for the custom integrand

We now generate the pedagogical visualization for the
manually defined integration problem.

The plot shows how the chosen quadrature rule samples
the integrand and how those sampled values contribute
to the numerical approximation of the integral.

```julia
plot_quadrature_coverage_1d(
    integrand, 
    bounds..., 
    ns;
    rule=rule,
    boundary=boundary,
    name=result_string,
    figs_dir=save_path,
    save_file=save_file
)
```

## Summary

In this notebook we visualized how quadrature rules interact
with an integrand.

We demonstrated two ways to construct the visualization:

1. Using a TOML configuration file
2. Defining the integrand and parameters directly in Julia

Such visualizations are useful for building intuition about
how quadrature methods approximate integrals and how
different rules sample the integration domain.
