# Maranatha.jl :: Pedagogical Visualization

This notebook demonstrates a pedagogical visualization of how
a quadrature rule samples an integrand over the integration domain.

Instead of focusing on convergence or extrapolation,
this visualization illustrates how quadrature nodes and weights
contribute to the integral estimate.

## Load package

Load the `Maranatha.jl` package.

```julia
using Maranatha
```

## Define integrand and quadrature setup

We define:

- the integrand
- the integration interval
- the quadrature rule
- the boundary condition
- the number of subintervals

These parameters determine how the quadrature nodes and weights
are placed over the domain.

```julia
integrand(x) = exp(-x) * cos(6x)
bounds = (0.0, π)
ns = 8
rule = :bspline_interp_p4
boundary = :LU_ININ
result_string = "demo"
save_file = true
```

## Visualize quadrature coverage

The function `plot_quadrature_coverage_1d` generates a pedagogical plot
showing:

- the integrand curve
- the quadrature nodes
- the contribution of each node to the numerical integral

This visualization helps illustrate how the selected quadrature rule
samples the function over the domain.

```julia
plot_quadrature_coverage_1d(
    integrand, 
    bounds..., 
    ns;
    rule=rule,
    boundary=boundary,
    name=result_string,
    save_file=save_file
)
```
